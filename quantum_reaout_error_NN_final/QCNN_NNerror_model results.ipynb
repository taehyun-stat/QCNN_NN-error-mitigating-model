```python
import matplotlib.pyplot as plt
import numpy as np

x_coor = np.arange(len(loss_avg_1))

fig = plt.figure()

plt.plot(x_coor, loss_avg_0, 'blue', label='model_0_loss')
plt.fill_between(x_coor, loss_avg_0-loss_std_0, loss_avg_0+loss_std_0, color='blue', alpha=0.2)

plt.plot(x_coor, loss_avg_1, 'orange', label='model_1_loss')
plt.fill_between(x_coor, loss_avg_1-loss_std_1, loss_avg_1+loss_std_1, color='orange', alpha=0.2)

plt.plot(x_coor, loss_avg_2, 'green', label='model_2_loss')
plt.fill_between(x_coor, loss_avg_2-loss_std_2, loss_avg_2+loss_std_2, color='green', alpha=0.2)

plt.plot(x_coor, loss_avg_3, 'red', label='model_3_loss')
plt.fill_between(x_coor, loss_avg_3-loss_std_3, loss_avg_3+loss_std_3, color='red', alpha=0.1)

plt.legend(loc='center right')
plt.title('MSE Loss in each QCNN_NNerror model \n for each iteration')
plt.xlabel('iteration')
plt.ylabel('loss')
# plt.ylim([0, 0.5])

plt.figure(figsize=(12,10))
plt.rcParams["figure.figsize"] = (12,10)
plt.rcParams['font.size'] = '18'

fig.savefig('Result/Qnode_NNerror_model_loss.jpg', dpi=400)
plt.show()
```
